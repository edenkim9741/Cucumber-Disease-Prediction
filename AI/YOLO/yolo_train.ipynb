```python
from ultralytics import YOLO

# 사전 학습된 yolov11n.pt 모델을 로드합니다.
model = YOLO('yolo11n.pt')
model.to('cuda')
# 이하 설정은 yolo12n, yolov10s에도 동일 적용
# data: 데이터셋 설정 파일 경로
# epochs: 전체 데이터셋을 몇 번 반복하여 학습할지 결정
# batch: 한 번에 몇 개의 이미지를 처리할지 결정 (GPU 메모리에 따라 조절)
# imgsz: 학습에 사용할 이미지 크기
# patiende: 지정 횟수만큼 연속으로 모델 개선이 없으면 조기 종료할지 결정
# lr0: 기본 learning rate, deafult 값은 0.01
results = model.train(data='D:/jiseong/cucumber/dataset/data.yaml', epochs=200, batch=32, imgsz=640, patience=50, workers=8)
```

```python
# 객체 탐지 및 바운딩 박스 체크 후 저장
model = YOLO('D:/jiseong/cucumber/runs/detect/train4/weights/best.pt') # 훈련된 Yolo11n
result = model.predict(
    source='D:/jiseong/cucumber/test_dataset/images/test',
    save=True,      # 각 이미지의 예측 결과를 저장하게 합니다.
    conf=0.25,       # 신뢰도 25% 이상인 객체만 표시
    save_txt=True   # 바운딩 박스 좌표를 .txt 파일로도 저장
)
```

```python
# Test 데이터셋을 사용하여 모델의 최종 성능을 평가합니다.
metrics = model.val(data='D:/jiseong/cucumber/test_dataset/test_data.yaml',split='test') # 'test' 세트를 사용하라고 명시
```
